```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
#
plt.style.use('helvet')
#
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(6,2.5))

g1 = np.loadtxt('../dummy_data/done/2.23_igdl_ssf.68573-68580')
g2 = np.loadtxt('../dummy_data/done/2.27_igdl_ssf.68573-68580')
g3 = np.loadtxt('../dummy_data/done/2.32_igdl_ssf.68573-68580')
ax[0].plot(g1[:,0], g1[:,1]*1e9, label='$\mathrm{2.25\,cm}$')
ax[0].plot(g2[:,0], g2[:,1]*1e9, label='$\mathrm{2.27\,cm}$')
ax[0].plot(g3[:,0], g3[:,1]*1e9, label='$\mathrm{2.32\,cm}$')
ax[0].legend()
ax[0].set_xlim(0,52)
ax[0].set_ylim(bottom=0)
ax[0].set_xlabel('$\mathrm{F_{probe}\,[GHz]}$')
ax[0].set_ylabel('$\mathrm{Group\,Delay\,[ns]}$')
ax[0].legend(frameon=False)
ax[0].text(0.9, 0.15, 'a)', transform=ax[0].transAxes)

#
p1 = np.loadtxt('../dummy_data/done/2.23_pfl_ssf_cut.68573-68580')
p2 = np.loadtxt('../dummy_data/done/2.27_pfl_ssf_cut.68573-68580')
p3 = np.loadtxt('../dummy_data/done/2.32_pfl_ssf_cut.68573-68580')
ax[1].plot(p1[:,0], p1[:,1], label='$\mathrm{2.25\,cm}$')
ax[1].plot(p2[:,0], p2[:,1], label='$\mathrm{2.27\,cm}$')
ax[1].plot(p3[:,0], p3[:,1], label='$\mathrm{2.32\,cm}$')
ax[1].legend()
ax[1].legend(frameon=False)
ax[1].set_ylim(bottom=0)
ax[1].set_xlabel('$\mathrm{R\,[m]}$')
ax[1].set_ylabel('$\mathrm{Density\,[10^{19}m^{-3}]}$')
ax[1].set_xticks([2.0,2.05,2.1,2.15,2.2, 2.25])
ax[1].set_xlim(1.97, 2.28)
ax[1].axvline(2.15, ls='--', color='k')
ax[1].axvspan(1.6, 2.15, alpha=0.2, color='k')
ax[1].axvspan(2.22, 2.3, alpha=0.2, color='b')
ax[1].text(2.25, 1.0, 'Limiter\nshadow', rotation=90, horizontalalignment='center')
ax[1].text(2.13, 0.4, 'Sep.', rotation=90)
ax[1].text(0.05, 0.15, 'b)', transform=ax[1].transAxes)

plt.tight_layout()
plt.savefig('Expinit.png', dpi=300)
plt.show()
```
